```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dane = pd.read_csv("creditcard.csv")
dane = dane.drop("Class", axis=1)
```

```python
pca = PCA(n_components = 30)
stdScaler=StandardScaler()
```

```python
scaledDf=stdScaler.fit_transform(dane)
pca.fit_transform(scaledDf)
print(sum(pca.explained_variance_ratio_))
```

```python
component_number=np.arange(30)
total_variance_ratios=[]
for n in component_number:
    pca=PCA(n_components=n)
    pca.fit_transform(scaledDf)
    total_variance_ratios+=[sum(pca.explained_variance_ratio_)]
```

```python
plt.figure(figsize=(4,4),dpi=150)
plt.grid()
plt.plot(component_number,total_variance_ratios,marker='o')
plt.axhline(y = 0.9, color = 'r')
plt.xlabel("Liczba cech")
plt.ylabel("Całkowity odsetek wariancji")
```
